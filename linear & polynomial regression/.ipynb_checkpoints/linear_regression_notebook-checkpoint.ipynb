```python
#Task 1.a
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load data from files
train_data = pd.read_csv('trainProject2.txt', header=None, names=['X', 'Y'])
test_data = pd.read_csv('testProject2.txt', header=None, names=['X', 'Y'])

# Extracting X and Y from our datasets
X_train = train_data[['X']].to_numpy()
y_train = train_data['Y'].to_numpy()

X_test = test_data[['X']].to_numpy()
y_test = test_data['Y'].to_numpy()

# Setting the frequency increment k
k = 0.5

def generate_features(X, k, depth):
    """
    This function generates features for each depth level.
    """
    features = np.ones(shape=(X.shape[0], 1))

    for i in range(1, depth + 1):
        # Generate the feature sin(i * k * X) and append it to the feature matrix
        feature = np.sin(i * k * X)
        features = np.concatenate((features, feature), axis=1)

    return features

# Define the maximum depth to consider
max_depth = 3

# Storage for MSE of models with different depths
mse_values = []

# Train models for different depths
for depth in range(1, max_depth + 1):
    # Generate features based on the current depth
    X_train_features = generate_features(X_train, k, depth)
    X_test_features = generate_features(X_test, k, depth)
    # Create and train a linear regression model with the current features
    model = LinearRegression()
    model.fit(X_train_features, y_train)
    # Make predictions and calculate the error
    predictions = model.predict(X_test_features)
    mse = mean_squared_error(y_test, predictions)
    mse_values.append(mse)

    print(f"Depth: {depth},  MSE: {mse}")

# After training all models, you might want to select the model with the lowest MSE
best_depth = np.argmin(mse_values) + 1  # adding 1 because depths start from 1 in the loop
print(f"The best model is at depth {best_depth} with an MSE of {mse_values[best_depth - 1]}")
```

```python
#Task 1.b
import numpy as np
import matplotlib.pyplot as plt

def load_data(train_file, test_file):
    train_data = np.loadtxt(train_file, delimiter=',')
    test_data = np.loadtxt(test_file, delimiter=',')
    return train_data, test_data

def prepare_features(x, depth):
    X = np.ones((len(x), 1))
    for i in range(1, depth + 1):
        X = np.hstack((X, x ** i))
    return X

def linear_regression(X, y):
    XT_X_inv = np.linalg.inv(X.T @ X)
    XTy = X.T @ y
    theta = XT_X_inv @ XTy
    return theta

def predict(X, theta):
    return X @ theta

def plot_results(x, y, theta, depth):
    plt.scatter(x, y, color='red', marker='o', label='Data points')
    x_values = np.linspace(min(x), max(x), 400).reshape(-1, 1)
    X_values = prepare_features(x_values, depth)
    predictions = predict(X_values, theta)
    plt.plot(x_values, predictions, color='blue', label=f'Fit (degree {depth})')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Polynomial Regression')
    plt.legend()
    plt.show()

def main():
    train_file = 'trainProject2.txt'
    test_file = 'testProject2.txt'

    # Load data
    train_data, test_data = load_data(train_file, test_file)
    x_train, y_train = train_data[:, 0].reshape(-1, 1), train_data[:, 1]
    x_test, y_test = test_data[:, 0].reshape(-1, 1), test_data[:, 1]

    max_depth = 7
    mse_values = []

    for depth in range(max_depth + 1):
        X_train = prepare_features(x_train, depth)
        X_test = prepare_features(x_test, depth)

        # Train model
        theta = linear_regression(X_train, y_train)

        # Make predictions on the test set
        predictions = predict(X_test, theta)

        # Calculate the Mean Squared Error on the test set
        mse = np.mean((y_test - predictions) ** 2)
        print(f"Degree {depth} Test Mean Squared Error:", mse)
        mse_values.append(mse)

        # Plotting the results
        plot_results(x_train, y_train, theta, depth)

    # Identify the best degree
    best_degree = np.argmin(mse_values)
    print(f"\nThe best model is with degree {best_degree}, with a Mean Squared Error of {mse_values[best_degree]}")

if __name__ == "__main__":
    main()
```

```python
#Task 1.c
import numpy as np
import matplotlib.pyplot as plt

# Load the data from text file
def load_data(file_path):
    data = np.loadtxt(file_path, delimiter=',')
    X, y = data[:, 0], data[:, 1]
    return X, y

X_train, y_train = load_data('trainProject2.txt')
# X_test, y_test = load_data('testProject2.txt')  # Uncomment if you have test data

# Polynomial Regression
def polynomial_regression(X, y, degree):
    # Add bias and polynomial features
    X_poly = np.ones((len(X), 1))
    for d in range(1, degree + 1):
        X_poly = np.hstack((X_poly, X.reshape(-1, 1) ** d))

    # Use the Normal Equation method to find the best fit
    theta = np.linalg.inv(X_poly.T.dot(X_poly)).dot(X_poly.T).dot(y)
    return theta

def plot_polynomial_regression(X, y, theta, degree):
    plt.scatter(X, y, color='red', marker='o', label='Data points')

    # Creating a continuous line for the polynomial regression result
    X_values = np.linspace(min(X), max(X), 400).reshape(-1, 1)
    X_poly_values = np.ones((len(X_values), 1))
    for d in range(1, degree + 1):
        X_poly_values = np.hstack((X_poly_values, X_values ** d))

    predictions = X_poly_values.dot(theta)

    plt.plot(X_values, predictions, color='blue', label=f'Degree {degree} fit')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Polynomial Regression')
    plt.legend()
    plt.show()

# Analyzing models with varying degrees
degrees = [1, 2, 3]  # you can specify the degrees you want to check

for degree in degrees:
    theta = polynomial_regression(X_train, y_train, degree)
    plot_polynomial_regression(X_train, y_train, theta, degree)
```

```python
#Task 1.d
import numpy as np
import matplotlib.pyplot as plt

# Load the data
train_data = np.loadtxt('trainProject2.txt', delimiter=',')
test_data = np.loadtxt('testProject2.txt', delimiter=',')

X_train = train_data[:, 0].reshape(-1, 1)
y_train = train_data[:, 1]
X_test = test_data[:, 0].reshape(-1, 1)
y_test = test_data[:, 1]

# Function to transform input data into polynomial form
def polynomial_features(X, degree):
    X_poly = np.ones((len(X), 1))
    for i in range(1, degree + 1):
        X_poly = np.hstack((X_poly, np.power(X, i)))
    return X_poly

# Function to perform linear regression
def linear_regression(X, y):
    # Normal Equation: (X^T * X)^-1 * X^T * y
    theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    return theta

# Analyzing models with degrees from 1 to 9 and storing their performance metrics
degrees = range(1, 10)
errors = []

for degree in degrees:
    # Transforming the original features into polynomial features of specified degree
    X_poly_train = polynomial_features(X_train, degree)
    X_poly_test = polynomial_features(X_test, degree)

    # Performing linear regression on the polynomial features
    theta = linear_regression(X_poly_train, y_train)
    y_test_predict = X_poly_test.dot(theta)

    # Calculating the Mean Squared Error (MSE)
    mse = ((y_test - y_test_predict) ** 2).mean()
    errors.append(mse)

# Identifying the degree that yields the minimum error
optimal_degree = degrees[np.argmin(errors)]

# Plotting the error results
plt.figure(figsize=(10, 6))
plt.plot(degrees, errors, marker='o', linestyle='-', color='b', label='Mean Squared Error')
plt.axvline(x=optimal_degree, color='r', linestyle='--', label=f'Min Error at Degree {optimal_degree}')
plt.title('Mean Squared Error for Various Polynomial Degrees')
plt.xlabel('Degree of Polynomial')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Displaying the optimal degree
print(f"The optimal degree of the polynomial is {optimal_degree} with the lowest mean squared error {errors[optimal_degree-1]}")
```
